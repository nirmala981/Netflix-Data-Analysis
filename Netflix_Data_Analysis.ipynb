```python
#IMPORT LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
LOAD DATASET
```

```python
#LOAD DATASET 
df = pd.read_csv("netflix_titles.csv")

df.head()
```

```python
DATA CLEANING
```

```python
#DATA INFO IN ROWS AND COLUMNS
df.shape
```

```python
#COLUMNS
df.info()
```

```python
#MISSING VALUES
df.isnull().sum()
```

```python
#FILLING MISSING VALUES
df['country'].fillna("Unknown", inplace=True)
df['rating'].fillna("Unknown", inplace=True)
df['director'].fillna("Unknown", inplace=True)
```

```python
DATA VISUALIZATION
```

```python
# 1.MOVIES AND TV SHOWS
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
# 2.TOP 10 CONTRIES PRODUCING CONTENT
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar')
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Shows")

plt.show()
```

```python
#3. CONTENT RELEASED PER YEAR
df['release_year'].value_counts().sort_index().plot()
plt.title("Content Released Per Year")
plt.xlabel("Year")
plt.ylabel("Count")

plt.show()
```

```python
# 4.TOP GENRES ON NETFLIX
genres = df['listed_in'].str.split(',', expand=True).stack()

top_genres = genres.value_counts().head(10)

top_genres.plot(kind='bar')

plt.title("Top Genres on Netflix")
plt.show()
```

```python
# 5.RATING DISTRIBUTION
df['rating'].value_counts().head(10).plot(kind='bar')

plt.title("Ratings Distribution")
plt.show()
```


# Final Insights

1.Netflix has significantly more Movies compared to Tv shows,indicating that the platform focuses more on movie content

2.The United states produces the highest number of Netfix titles ,followed by countries like India and the United Kingdom.

3.The number of shows released on Netflix has increased rapidly after 2015, showing the platforms expansion in recent years.
